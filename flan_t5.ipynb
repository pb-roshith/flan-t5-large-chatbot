```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

tokenizer = AutoTokenizer.from_pretrained("google/flan-t5-large")
model = AutoModelForSeq2SeqLM.from_pretrained("google/flan-t5-large")
```

```python
inp = input("Enter your query : ")
inputs = tokenizer(inp, return_tensors="pt")
outputs = model.generate(**inputs)
print(tokenizer.batch_decode(outputs, skip_special_tokens=True))
```

```python
ask = input('Enter your query : ')
inputs = tokenizer(ask, return_tensors="pt")
outputs = model.generate(**inputs, min_length=50, max_new_tokens=90, \
                         length_penalty=2, num_beams=8, \
                         no_repeat_ngram_size=2, early_stopping=True)
outputs_text_flan_t5 = tokenizer.batch_decode(outputs, skip_special_tokens=True)
print(outputs_text_flan_t5)
```
